# **Task 1: TITANIC SURVIVAL PREDICTION**

### Importing all necessary libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

### Importing the csv file

```python
df = pd.read_csv('Titanic.csv')
```

### Checking the dataframe's info

```python
df.info()
```

### Dataframe statistics

```python
df.describe().T
```

### Dividing the dataframe into two seperate dataframes: survived & died

```python
survived = df[df['Survived'] == 1]
died = df[df['Survived'] == 0]
```

### Checking our dataframes

```python
survived.info()
died.info()
```

### Limiting to our need

```python
survived = survived[['PassengerId', 'Name', 'Age', 'Sex']]
died = died[['PassengerId', 'Name', 'Age', 'Sex']]
```

### Checking Again

```python
survived.info()
died.info()
```

### Ploting the count of each category (died = 0 / survived = 1)

```python
sns.countplot(data = df, x = 'Survived')
```

### Ploting the count of survival people grouped by sex

```python
sns.countplot(data = survived, x = 'Sex')
```

## As a conclusion, all survived people were women and no men were survived.
